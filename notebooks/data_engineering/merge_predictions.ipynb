```python
import pandas as pd
import sys
import os
```

```python
dirpath = "/nfs/turbo/umms-indikar/shared/projects/MC3R/hypomap/predictions/"
outpath = "/nfs/turbo/umms-indikar/shared/projects/MC3R/hypomap/merged_predictions.csv"

df_list = []

for f in os.listdir(dirpath):
    fullpath = f"{dirpath}{f}"
    tmp = pd.read_csv(fullpath)
    name = f.replace("predictions_", "").replace(".csv", "")

    tmp = tmp[['index', 'predicted_celltype']]
    tmp = tmp.set_index('index')
    tmp.columns = [name]

    df_list.append(tmp)

    
    
df = pd.concat(df_list, ignore_index=False, axis=1)
df = df.reset_index(drop=False)

df.to_csv(outpath, index=False)
df.head()
```
